# Sample Quality Check

Use this notebook after running the pipeline to inspect final RAG chunks and validation output.

```python
from pathlib import Path
import json

root = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
chunks_path = root / 'data' / 'output' / 'ramayana_rag_chunks.jsonl'
validation_path = root / 'data' / 'output' / 'validation_report.json'

chunks = []
if chunks_path.exists():
    chunks = [json.loads(line) for line in chunks_path.read_text(encoding='utf-8').splitlines() if line.strip()]

print('chunks:', len(chunks))
if chunks:
    sample = chunks[0]
    print(sample['chunk_id'])
    print(sample['english_translation'][:800])

if validation_path.exists():
    print(json.dumps(json.loads(validation_path.read_text(encoding='utf-8')), indent=2)[:2000])
```
